# III. Motivasyon ve Arka Plan 

## Veri Setinin Seçim Nedeni

Meme kanseri, dünya genelinde kadınlar arasında en yaygın kanser türlerinden biridir ve erken teşhis, hastalığın başarılı bir şekilde tedavi edilmesinde kritik öneme sahiptir. Breast Cancer Prediction Dataset, meme kanserinin tanısına yardımcı olabilecek çeşitli biyopsi sonuçlarını içerir ve bu veriler üzerinde çalışmak, makine öğrenimi modelleri kullanarak kanser teşhisinde doğruluğu artırabilir. Bu nedenle, meme kanseri verisetini seçmemizin başlıca nedenleri şunlardır:
  
- Meme kanseri, tıbbi alanda önemli bir sağlık sorunu olup, bu konuyla ilgili veri setleri üzerinde çalışmak gerçek dünya problemlerine çözümler geliştirmeye katkı sağlar.
-  Veri seti, meme kanseri teşhisi için kullanılabilecek çeşitli özellikleri içerir (örneğin, ortalama yarıçap, ortalama doku, ortalama çevre, ortalama alan, ortalama düzgünlük gibi). Bu özellikler, kanserli ve kansersiz hücreler arasındaki farkları anlamamıza yardımcı olur.
- : Veri seti, makine öğrenimi algoritmaları için uygun yapıda olup, sınıflandırma problemleri üzerinde çalışmak için iyi bir örnek teşkil eder.tir.


## Hesapladığımız Değerler ve Nedenleri

Meme kanseri teşhisinde kullanılan verilerle, aşağıdaki hesaplamaları ve analizleri yaparız:

**Özelliklerin Analizi:**

- **Ortalama Yarıçap (mean_radius):** Hücrelerin boyutu ve büyüme hızı hakkında bilgi verir. Daha büyük yarıçap değerleri, kanserli hücrelerin daha agresif olabileceğini veya tümör oluşumunun daha ileri bir aşamada olduğunu gösterebilir.

- **Ortalama Doku (mean_texture):** Meme dokusundaki doku değişimi veya heterojenite, kanser hücrelerinin farklılıklarını ve kanserli dokunun yayılma potansiyelini yansıtabilir.

- **Ortalama Çevre (mean_perimeter):** Kanserli dokunun yayılma ve büyüme hızı hakkında bilgi verir. Daha büyük çevre değerleri, kanserli dokunun yayılma potansiyelini gösterebilir.

- **Ortalama Alan (mean_area):** Kanserli dokunun boyutu ve büyüme hızıyla ilgili bilgi verir. Daha büyük alan değerleri, kanserli dokunun daha büyük ve daha agresif olabileceğini gösterebilir.

- **Ortalama Düzleşme (mean_smoothness):** Meme dokusundaki düzleşme derecesi, hücrelerin homojenliği ve düzensizliği hakkında bilgi verir. Daha düşük düzleşme değerleri, kanserli hücrelerin daha hızlı büyüyebileceğini veya farklılıklar gösterebileceğini işaret edebilir.


**Makine Öğrenimi Modelleriyle Tahmin:**


- **Model Eğitimi ve Testi:** Veri setindeki özellikleri kullanarak çeşitli makine öğrenimi algoritmaları (örneğin, lojistik regresyon, karar ağaçları, rastgele ormanlar, destek vektör makineleri) ile modeller oluştururuz.

- **Model Karşılaştırması:** Farklı modellerin performansını doğruluk, hassasiyet, geri çağırma, F1 skoru gibi metriklerle değerlendiririz. Bu karşılaştırma, en iyi performans gösteren modeli seçmemizi sağlar.

- **En İyi Modelin Belirlenmesi:** En yüksek doğruluk oranına sahip modeli belirleriz. Genellikle SVM (Destek Vektör Makineleri) gibi modeller, belirli veri setlerinde daha iyi performans gösterebilir.

## Neden Bu Hesaplamaları Yapıyoruz?

- Farklı makine öğrenimi modelleri ile çalışmak, hangi modelin belirli bir veri seti için en iyi performansı gösterdiğini anlamamıza yardımcı olur. Bu, gelecekteki çalışmalar ve uygulamalar için yol gösterici olur.

- Meme kanseri ile ilişkili özelliklerin anlaşılması, kanserin biyolojik özellikleri hakkında daha fazla bilgi edinmemizi sağlar ve bu bilgiler, kanser araştırmalarına katkıda bulunur.

# IV. Özellik seçimi

Özellik seçimi, bir makine öğrenimi veya veri analizi problemi için en uygun özellikleri seçme veya oluşturma sürecidir. Bu süreç, modelin performansını artırmak, gereksiz veya gürültülü özellikleri ortadan kaldırmak ve modelin genelleme yeteneğini iyileştirmek amacıyla gerçekleştirilir.

Özellik seçimi için çeşitli yöntemler ve yaklaşımlar vardır:

- Görsel Analiz: Veri setinin görsel analizi, özellikler arasındaki ilişkileri ve özelliklerin hedef değişkenle olan ilişkisini anlamak için kullanılır. Bu, önemli özellikleri belirlemeye yardımcı olabilir.

- Korelasyon Analizi: Özellikler arasındaki ilişkiyi ölçmek için korelasyon matrisi kullanılır. Yüksek korelasyona sahip olan özellikler birlikte kullanılmamalıdır.

- Özellik Önemi Sıralaması: Bazı makine öğrenimi modelleri, özelliklerin önem derecelerini belirlemek için bir önem sıralaması sağlar. Özelliklerin önem dereceleri, modelin performansını artırmak için önemli bir rehber olabilir.

- Model Tabanlı Seçim: Özelliklerin performans üzerindeki etkisini doğrudan değerlendirmek için model tabanlı yöntemler kullanılabilir. Örneğin, ağaç tabanlı yöntemler özellik önemlilik sıralamalarını sağlayabilir.

- RFE (Recursive Feature Elimination): RFE, model performansını maksimize edecek şekilde özelliklerin alt kümesini seçer. Başlangıçta tüm özellikler kullanılır, ardından en az öneme sahip özellikler teker teker çıkarılır ve modelin performansı değerlendirilir.

- PCA (Principal Component Analysis): PCA, özelliklerin boyutunu azaltmak için kullanılır. Veri setindeki özellikler arasındaki varyansı en iyi şekilde koruyarak, özelliklerin azaltılmış bir alt kümesini oluşturur.

Özellik seçimi, modelin performansını iyileştirmek ve overfitting riskini azaltmak için önemlidir. Hangi yöntemin kullanılacağı, veri setinin özelliklerine, problem alanına ve mevcut kaynaklara bağlı olacaktır.

```python
from sklearn.decomposition import PCA

# PCA modelini oluştur
pca = PCA()

# Veri setini PCA'ya uygula ve bileşenleri bul
pca.fit(veri)

# Bileşenleri al
bilesenler = pca.components_

print(bilesenler)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Veri setini yükle
veri = pd.read_csv("C:/Users/dunyadanbus/Desktop/Breast_cancer_data.csv")

# Bağımsız değişkenler ve hedef değişkeni ayır
X = veri.drop('diagnosis', axis=1)
y = veri['diagnosis']

# Rastgele Orman sınıflandırıcı modelini oluştur
rf_model = RandomForestClassifier()

# Modeli eğit
rf_model.fit(X, y)

# Özellik önemlerini al
ozellik_onemleri = rf_model.feature_importances_

# Özellik önemlerini DataFrame'e dönüştür
ozellik_onemleri_df = pd.DataFrame({'Özellikler': X.columns, 'Önem Derecesi': ozellik_onemleri})

# Özellik önemlerini önem derecesine göre sırala
ozellik_onemleri_df = ozellik_onemleri_df.sort_values(by='Önem Derecesi', ascending=False)

# Sıralanmış özellik önemlerini ekrana yazdır
print(ozellik_onemleri_df)
```

- mean_perimeter özelliği, mean_areasonra gelen en önemli ikinci özelliktir.

# VI. Algoritma seçimi

- **Destek Vektör Makineleri(SVM):**
SVM, sınıflandırma ve regresyon problemlerinde kullanılabilen güçlü bir algoritmadır.
Veri kümemizdeki özellikler arasındaki doğrusal veya doğrusal olmayan ilişkileri belirlemek için idealdir.
SVM, yüksek boyutlu veri kümeleriyle iyi çalışır ve sınırlayıcı hiperparametreler sayesinde overfitting'e karşı dirençlidir.

- **Karar Ağaçları (Decision Trees):**
Karar ağaçları, basitlikleri ve yüksek yorumlanabilirlikleri nedeniyle tercih edilen bir algoritmadır.
Özellikler arasındaki etkileşimleri ve ilişkileri anlamak için kolayca görselleştirilebilirler.
Karar ağaçları, eksik veri ve aykırı değerlerle iyi başa çıkabilir ve genellikle karmaşık olmayan veri yapıları için uygundur.

- **K-En Yakın Komşu (KNN) Algoritması:**
KNN, örnek tabanlı bir algoritma olup, sınıflandırma ve regresyon problemlerinde kullanılabilir.
KNN, basit ve etkili bir algoritmadır ve eğitim aşaması gerektirmez. Modeli eğitmek yerine, yeni veri noktalarını sınıflandırmak için kullanılır.
Veri setindeki benzer örnekleri bulmak ve onlara dayanarak sınıflandırma yapmak için idealdir.

- **Lojistik Regresyon (Logistic Regression):**
Lojistik regresyon, sınıflandırma problemlerinde yaygın olarak kullanılan bir algoritmadır.
Basitlik ve yorumlanabilirlik açısından avantaj sağlar.
Lojistik regresyon, bağımsız değişkenlerin sınıflandırılmış bir çıktıya olan etkisini anlamak için katsayıları kullanır.

- **Rastgele Orman (Random Forest):**
Rastgele ormanlar, karar ağaçlarının birleşiminden oluşan bir enseble yöntemidir.
Veri setindeki özellikler arasındaki ilişkileri ve önem derecelerini belirlemek için etkili bir yöntemdir.
Rastgele ormanlar, overfitting'e karşı dirençlidir ve genellikle yüksek boyutlu ve gürültülü veri kümeleriyle iyi çalışırlar.
